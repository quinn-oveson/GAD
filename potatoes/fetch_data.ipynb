```python
import requests
import pandas as pd
```

```python

API_KEY = "YOUR_API_KEY_HERE"

url = "https://quickstats.nass.usda.gov/api/api_GET/"

params = {
    "key": API_KEY,

    # Commodity
    "commodity_desc": "POTATOES",

    # What variable to retrieve
    "statisticcat_desc": "YIELD",

    # Units
    "unit_desc": "CWT / ACRE",

    # Geography
    "state_name": "IDAHO",
    "agg_level_desc": "COUNTY",

    # Frequency
    "freq_desc": "ANNUAL",

    # Pull all years available
    "year__GE": "1900",

    # Output format
    "format": "JSON"
}

response = requests.get(url, params=params)

data = response.json()["data"]

df = pd.DataFrame(data)

# Keep useful columns
cols = [
    "year",
    "county_name",
    "Value",
    "short_desc"
]

df = df[cols]

# Clean yield values
df["Value"] = (
    df["Value"]
    .str.replace(",", "", regex=False)
    .astype(float)
)

# Sort
df = df.sort_values(["county_name", "year"])

print(df.head())

# Save
df.to_csv("idaho_potato_yield.csv", index=False)
```
